```python
import json
import pandas as pd
import matplotlib.pyplot as plt
import bar_chart_race as bcr
```

```python
tm_json_file = 'ticket_master_attractions_data.json'
tm_json_file
```

```python
tm_json = pd.read_json(tm_json_file)
tm_json
sort_tm_json = tm_json.sort_values(by='attraction_id', ascending=True)
sort_tm_json
```

```python
tm_df = tm_json[['attraction_name',
                      'segment', 'genre', 
                      'sub_genre', 
                      'event_name', 
                      'venue_city', 
                      'venue_state', 
                      'venue_country',
                      'venue_longitude',
                      'venue_latitude',
                      'start_date']]
tm_df
```

```python
sort_tm_df = tm_df.sort_values(by='attraction_name', ascending=True)
sort_tm_df
```

```python
filtered_tm_df = sort_tm_df[['attraction_name', 'genre', 'event_name', 'start_date']]
filtered_tm_df.dtypes
```

```python
unique_attractions = filtered_tm_df['attraction_name'].unique()
unique_attractions
```

```python
bcr_tm_df = filtered_tm_df[['start_date','attraction_name', 'genre', 'event_name']]
bcr_tm_df
```

```python
df_no_dups = bcr_tm_df.drop_duplicates(subset=['start_date', 'attraction_name', 'event_name'])
df_no_dups
```

```python
df_no_dups = df_no_dups.sort_values(by='start_date', ascending=True)
df_no_dups
```

```python
cleaned_bcr_df = df_no_dups[['start_date', 'genre']]
cleaned_bcr_df
```

```python
cleaned_bcr_df = cleaned_bcr_df.reset_index(drop = True)
cleaned_bcr_df.head(20)
```

```python
cleaned_bcr_df['summed_count'] = cleaned_bcr_df.groupby(['start_date', 'genre']).transform('size')
```

```python
cleaned_bcr_df = cleaned_bcr_df.drop_duplicates(['start_date', 'genre']).reset_index(drop=True)
cleaned_bcr_df
```

```python
cleaned_bcr_df['total_count'] = cleaned_bcr_df.groupby('genre')['summed_count'].cumsum()
```

```python
cleaned_bcr_df.head(20)
```

```python
cleaned_bcr_df = cleaned_bcr_df[['start_date', 'genre', 'total_count']]
cleaned_bcr_df
```

```python
cleaned_bcr_df['start_date'] = pd.to_datetime(cleaned_bcr_df['start_date'])
cleaned_bcr_df.dtypes
```

```python
pivot_df = cleaned_bcr_df.pivot(index='start_date', columns='genre', values='total_count')

pivot_df.head(20)
```

```python
df_filled = pivot_df.fillna(method='ffill')
df_filled
```

```python
# Create the bar chart race
bcr.bar_chart_race(
    df=df_filled,

    # Assign name to gif
    filename='bar_chart_race.gif',
    # Title of graph
    title='How Many Upcoming Shows?',
    # Make orientation horizontal
    orientation='h',
    #Sort in descending order so highest value remains at the top
    sort='desc',
    # How many bars will be displayed?
    n_bars=10,
    # Make it so bars change positions as values change
    fixed_order=False,
    # Filter column colors
    filter_column_colors = True,
    # Number of steps between each year
    steps_per_period=5,
    #Length of each period (milliseconds)
    period_length=250,
    # Make interpolate False
    interpolate_period=False
)
```
